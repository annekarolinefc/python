# Arquivos

## Criar Arquivo

```python
import random

arq = open("numeros.txt", "w")

for i in range(0, 100):
    val = random.random()
    arq.write(str(val)+"\n")
print("Arquivo Criado")

arq.close()
```

## Lê Arquivo

open(file, mode = 'r')

Os modos são:
r - leitura
w - escrita


```python
arq = open("numeros.txt", "r")

for line in arq.readlines():
    print(float(line)*10)
    
arq.close()
```

outra forma:

```python
with open("numeros.txt", "r") as f:
    for line in f.readlines():
        print(float(line)*10)
```
